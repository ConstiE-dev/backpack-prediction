quick preprocesssing using fastai

```python
from fastai.tabular.all import *
import pandas as pd

train_data_original = pd.read_csv("dataset/train.csv")
test_data_original = pd.read_csv("dataset/test.csv")
```

```python
train_data = train_data_original.drop("id", axis=1)
```

```python
train_data
```

```python
splits = RandomSplitter(valid_pct=0.2)(range_of(train_data))
to = TabularPandas(train_data, procs=[Categorify, FillMissing,Normalize],
                   cont_names = ['Weight Capacity (kg)',], 
                   cat_names=['Brand','Material','Size','Compartments','Laptop Compartment','Waterproof','Style','Color'],
                   y_names='Price',
                   splits=splits)
```

```python
to.train.xs
```

```python
dls = to.dataloaders(bs=64)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=rmse)
learn.fit_one_cycle(1)

```

```python
to = TabularPandas(train_data, procs=[Categorify, FillMissing,Normalize],
                   cont_names = ['Weight Capacity (kg)',], 
                   y_names='Price',
                   splits=splits)
dls = to.dataloaders(bs=64)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=rmse)
learn.fit_one_cycle(1)
```

```python
learn = tabular_learner(dls, metrics=rmse, wd=0.1)
learn.fit_one_cycle(1)
```
